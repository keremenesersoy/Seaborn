```python
import seaborn as sns

data = sns.load_dataset('fmri')
df = data.copy()
df.head()
```

# lineplot

```python
sns.lineplot(x="timepoint" , y = "signal" , data = df);
# zaman noktalarına göre sinyallerin değişimi ile ilgili bize bir grafik verir
# koyu çizgi ortalama değere karşılık gelir
# etrafındaki açık renkli alanlar standart sapmalardır
```

### 3.boyut ekleme

```python
sns.lineplot(x="timepoint" , y = "signal" , data = df , hue="event" , style="event" , markers=True , dashes=False);
```

### 4.boyut ekeleme

```python
sns.lineplot(x="timepoint" , y = "signal" , data = df , hue="region" , style="event" , markers=True , dashes=False);
```
