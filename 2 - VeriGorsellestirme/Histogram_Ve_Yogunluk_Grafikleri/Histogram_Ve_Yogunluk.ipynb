```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

### Distplot

```python
sns.distplot(df['price'] , kde = False , bins = 10);
# distplot ile değerleri belirli bir ölçüde böler
# ve değişkenlerin hangi aralıkta oldugunu gösteren bir histogram verir
# bins argümanı kutuların sayısını belirtir ne kadar yüksekse histogram o kadar hassas olur
```

```python
sns.distplot(df['price'] , hist=False)
# eğer distplot fonksiyonuna direk olarak bir öznitelik verirsek
# y ekseninde ölçekler farklı olacağı için seaborn arka planda olasılık yoğunluk fonksiyonuna göre
# ölçeği indirgeyip bize yoğunluk grafiğide verir
# eğer histogramı istemiyorsak hist argümanına False girmemiz gerek
```

### kdeplot

```python
sns.kdeplot(df.price , shade=True) # kdeplot fonksiyonu yukardaki distplot fonksiyonunun hist = False olan halidir
# shade özelliğini True yaparsak grafiğin alt kısmını doldurur
```

```python
(sns.FacetGrid(df,
        hue = "cut", #cut özniteliğini yeni bir boyut olarak ekledik
        height=5, # yükseklik
        xlim = (0,10000)) # x ekseni 0 ile 1000 arasındaki değerleri alacağını belirttik
.map(sns.kdeplot ,"price" , shade=True ) #  pricea göre bir kdeplot grafiği oluşturup , bu grafiği yeni bir boyut olan cut ile
                    # eşlememiz için map kullandık
.add_legend()); # legend ile başlık etiketi ekledik
```
