```python
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
df = data.copy()
df.head()
```

### Scatterplot

```python
sns.scatterplot(x = "total_bill" , y = "tip" , data = df);
### scatter plot 2 değişken arasındaki ilişkiyi nokta şeklinde gösterir
```

### 3.boyut ekleme

```python
sns.scatterplot(x = "total_bill" , y = "tip" , hue = "time" , data = df);
```

### style özelliği

```python
sns.scatterplot(x = "total_bill" , y = "tip" , hue = "time" , data = df , style="time");
# style özelliğine değiken ismi verirsek farklı şekiller verir
```

```python
sns.scatterplot(x = "total_bill" , y = "tip" , hue = "day" , data = df , style = "day");
```

### size özelliği

```python
sns.scatterplot(x = "total_bill" , y = "tip" , data = df , size="size" , hue="size");
# hue argümanına da size yazarsak daha renkler soluktan koyuya doğru sıralanır
```
