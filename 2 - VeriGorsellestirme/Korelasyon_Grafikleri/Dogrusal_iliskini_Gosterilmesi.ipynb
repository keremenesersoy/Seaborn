```python
import seaborn as sns
import matplotlib.pyplot as plt 

data = sns.load_dataset("tips")
df = data.copy()
```

### lmplot

```python
sns.lmplot(x = "total_bill" , y = "tip" , data = df);
# 2 değişken arasındaki dogrusal ilişkiyi gösterir
# doğrunun etrafındaki açık renkli kısım sapmaları ifade eder 
```

### 3.boyut ekleme

```python
sns.lmplot(x = "total_bill" , y = "tip" , data = df , hue = "smoker");
# grafikte sigara içmeyenlerin daha fazla bahşiş bıraktığını görebiliriz
```

### col özelliği

```python
sns.lmplot(x = "total_bill" , y = "tip" , data = df , hue = "smoker" , col = "time");
# col argümanı ile özelliğe göre çeşitli grafikler elde ederiz
```

### row özelliği

```python
sns.lmplot(x = "total_bill" , y = "tip" , data = df , hue = "smoker" , row = "sex", col = "time");
# row argümanı ile özelliğe göre çeşitli grafikler elde eder
```
