```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
df = data.copy()
df.head()
```

```python
df.dtypes
```

### pairplot

```python
sns.pairplot(data=df)
# grafik bize 2 bilgi sunar 
# 1 - eksende yer alan grafikler -> veri seti ierisinde yer alan deikenlerin dalmlarn ifade eder
# 2 - eksenlerden yaplacak kesitirmeler ile deikenlerin arasndaki ilikinin sunulmas
```

### 3.Boyut Ekleme

```python
sns.pairplot(data = df , hue = "species" );
# 3.boyutu eklediimizde grafikler eitlenir
```

### markers zellii

```python
sns.pairplot(data = df , hue = "species" , markers = ["o" , "s" , "d"]);
# markers zellii ile iaretleri belirleriz
```

### Grafie doru ekleme

```python
sns.pairplot(data = df , kind = "reg" , hue = "species");
```
