```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
df = data.copy()
df.head()
```

```python
df.dtypes
```

### pairplot

```python
sns.pairplot(data=df)
# grafik bize 2 bilgi sunar 
# 1 - eksende yer alan grafikler -> veri seti içerisinde yer alan değişkenlerin dağılımlarını ifade eder
# 2 - eksenlerden yapılacak kesiştirmeler ile değişkenlerin arasındaki ilişkinin sunulması
```

### 3.Boyut Ekleme

```python
sns.pairplot(data = df , hue = "species" );
# 3.boyutu eklediğimizde grafikler çeşitlenir
```

### markers özelliği

```python
sns.pairplot(data = df , hue = "species" , markers = ["o" , "s" , "d"]);
# markers özelliği ile işaretleri belirleriz
```

### Grafiğe doğru ekleme

```python
sns.pairplot(data = df , kind = "reg" , hue = "species");
```
