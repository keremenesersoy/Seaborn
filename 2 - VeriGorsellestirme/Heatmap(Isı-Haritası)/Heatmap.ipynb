```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")
df = data.copy()
df.head()
```

```python
df.shape
```

```python
df['passengers'].describe()
```

```python
df = df.pivot("month" , "year" , "passengers")
```

### Heatmap

```python
sns.heatmap(df);
# ısı haritası veriyi temsil edip büyük resmi görmemizi sağlar
```

```python
sns.heatmap(df , annot=True , fmt="d" , linewidths = 0.5);
# linewidths hücrelerin arasının uzaklığını belirler
# fmt ve annot değerleri ise o hücredeki değeri göstermeyi sağlar
```

### Isı Haritası Bar Kaldırma

```python
sns.heatmap(df , annot=True , fmt="d" , linewidths = 0.5 , cbar=False);
# cbar özelliğini False yaparsak bar kalkar
```
