```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
cut_kategoriler = ['Fair' , 'Good' , 'Very Good' , 'Premium' , 'Ideal']
df['cut'] = df['cut'].astype(CategoricalDtype(categories=cut_kategoriler , ordered=True))
df.head()
```

### Cat Plot

```python
sns.catplot(x = 'cut' , y = 'price' , data=df) # kategorik değişkenleri görselleştirmek için kullanılır
```

### Barplot

```python
sns.barplot(x = 'cut' , y='price' , hue='color' , data=df) #hue argümanı ile aslında grafiğe bir boyut daha eklemiş oluyoruz
# hue argümanı verildiğinde 2 gruplaştırma işleminden sonra price değişkeninin ortalamasıdır grafiğin üstündeki çubuklar ise standart sapmadır
```

```python
df.groupby(['cut','color'])['price'].mean() # yukardaki grafikte hue değerine color verdiğimiz için
                    # cut ve colora göre grupladıktan sonra ortalamasını aldı bu değerleri manuel olarak yaptıgımızdada görebiliriz
```
