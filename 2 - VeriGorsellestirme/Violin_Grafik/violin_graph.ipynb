```python
import seaborn as sns
import pandas as pd

data = sns.load_dataset('tips')
df = data.copy()
```

```python
df.head()
```

### catplot

```python
sns.catplot(y = "total_bill" , kind = "violin" , data=df)
# ortadaki çizgi medianı ele alır
```

### 3.boyut ekleme

```python
sns.catplot(x = "day" , y = "total_bill" , kind = "violin" , hue = "sex" , data=df)
```
